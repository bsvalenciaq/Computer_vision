```python
r,c = img.shape
```

```python
sum_col=np.sum(img, axis = 0) #I add up all the elements for each column
sum_fil=np.sum(img, axis = 1) #I add up all the elements for each row
print(sum_fil, sum_col)
```

```python
img=np.delete(img, sum_col == r*255, axis=1) #Delete empty columns
img=np.delete(img, sum_fil == c*255, axis=0) #Delete empty rows
img.shape #result
```
