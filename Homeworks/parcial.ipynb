```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig=plt.figure(figsize=(20,8))
a1= plt.subplot2grid((6,10),(0,0),colspan=4,rowspan=2)
a2=plt.subplot2grid((6,10),(0,4), rowspan=5,colspan=2)
a3=plt.subplot2grid((6,10),(2,0),rowspan=3, colspan=4)
a4=plt.subplot2grid((6,10),(0,6),rowspan=5, colspan=4)#
#a6=plt.subplot2grid((4,7),(0,6),rowspan=4)
x=np.arange(1,10)

x1=np.linspace(0,2*np.pi,20)
y1=np.sin(x1)
y2=np.cos(x1)


l1=a3.plot(x1,y2,'bs-') # solid line with yellow colour and square marker
l1=a3.plot(x1,y1,'ro--') # dash line with green colour and circle marker
a3.legend(labels=('sin', 'cos'), loc='lower left') # legend placed atlower right
a3.axhline(y=1,color='grey',ls='--')
a3.axhline(y=0,color='grey',ls='--')
a3.axhline(y=-1,color='grey',ls='--')
a3.set_xticks([0,2*np.pi])
a3.set_yticks([-1,0,1])
#a3.set_xlabel(['0','$\pi$\'])
a3.set_xticklabels(['0','$2\pi$'])



data = [[30, 20, 50, 40],
 [40, 60, 50, 10],
 [10, 50, 10, 30]]
X = np.arange(4)
a1.bar(X + 0.00, data[0], color = 'y', width = 0.25)
a1.bar(X + 0.25, data[1], color = 'b', width = 0.25)
a1.bar(X + 0.50, data[2], color = 'r', width = 0.25)
a1.set_xticks([0.25,1.25,2.25,3.25])
a1.set_xticklabels([2015,2016,2017,2018])
a1.legend(labels=['audi','BWM','CHWEBROLET'])

x2=np.arange(1,11)

a2.plot(x2,np.exp(x2) ,'b-.')
a2.set_ylabel('exp(x)')
a2.set_yscale('log')
a5=a2.twinx()
a5.plot(x2,np.log(x2),'r--')
a5.set_ylabel('log(x)')
a5.set_yscale('linear')

xlist = np.linspace(-3.0, 3.0, 100)
ylist = np.linspace(-3.0, 3.0, 100)
X, Y = np.meshgrid(xlist, ylist)
Z = np.sqrt(X**2 + Y**2)
#fig,a4=plt.subplots(1,4)

cp = a4.contourf(X, Y, Z)
fig.colorbar(cp,ax=a4) # Add a colorbar to a plot
a4.set_title('$z=\sqrt{x^2+y^2}$')
a4.set_xlabel('X')
a4.set_ylabel('Y')

plt.tight_layout()

plt.show()
```
