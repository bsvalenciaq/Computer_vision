Crear imagenes de colores

```python
class Crear_imagen(): #Crear Clase
  """ 
 Clase con metodos para crear imagenes de color azul, verde, rojo, blanca, negra
  """
  def __init__(self, M): 
    self.M = M 
    x,y  = np.shape(self.M)
    self.x= x
    self.y = y

  def blue(self): #METODOS..
    
    return (plt.imshow(M, cmap = 'Blues', vmin = 0, vmax = 255, extent = (0,self.x, 0,self.y)),
  
    plt.title('Image blue'),
    plt.ylabel('Y'),
    plt.xlabel('X'))

  def green(self): #METODOS..
        
    return (plt.imshow(M, cmap = 'Greens', vmin = 0, vmax = 255, extent = (0,self.x, 0,self.y)),
 
    plt.title('Image green'),
    plt.ylabel('Y'),
    plt.xlabel('X'))

  def red(self): #METODOS..
    
    return (plt.imshow(M, cmap = 'Reds', vmin = 0, vmax = 255, extent = (0,self.x, 0,self.y)),
   
    plt.title('Image red'),
    plt.ylabel('Y'),
    plt.xlabel('X'))

  def gray(self): #METODOS..
    
    return (plt.imshow(M, cmap = 'Greys', vmin = 0, vmax = 255, extent = (0,self.x, 0,self.y)),
   
    plt.title('Imagen gray'),
    plt.ylabel('Y'),
    plt.xlabel('X')
    )
```

```python
M = np.random.randint(0,255, (1000,1000))
im = Crear_imagen(M)
blue = im.blue
green = im.green
red = im.red 
gray = im.gray
```
